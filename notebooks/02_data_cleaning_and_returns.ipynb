```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# load raw data
crypto_df = pd.read_csv("../Data/raw/crypto_prices_raw.csv")
crypto_df.head()
```

```python
crypto_df.info()
```

```python
crypto_df.isna().sum()
```

```python
crypto_df.duplicated().sum()
```

```python
# convert date column to datetime
crypto_df["Date"] = pd.to_datetime(crypto_df["Date"])
```

```python
crypto_df.dtypes
```

```python
# sort data
crypto_df = crypto_df.sort_values(by=["coin", "Date"]).reset_index(drop=True)
crypto_df.head()
```

```python
# calculate daily returns
crypto_df["daily_return"] = crypto_df.groupby("coin")["Close"].pct_change()
```

```python
crypto_df.head(10)
```

```python
# calculate cumulative returns
crypto_df["cumulative_return"] = (
    (1 + crypto_df["daily_return"])
    .groupby(crypto_df["coin"])
    .cumprod() - 1
)
```

```python
crypto_df.head(10)
```

```python
# summary stats daily returns
crypto_df.groupby("coin")["daily_return"].describe()
```
